##### 2 - Escreva uma função que encontre a mediana de uma lista de números reais ou inteiros. 

```
def CalcularMediana(lista):
  notas_ordenadas = sorted(lista)
  numero_nota= int(len(lista))
  if numero_nota % 2 ==0:
    return (notas_ordenadas[int(numero_nota/2)]+ notas_ordenadas[int(numero_nota/2-1)])/2.0
  else:
    return notas_ordenadas[int(numero_nota/2)]
```

```
print("Mediana = {:0.2f}".format(CalcularMediana(listaNotas)))
```

```
print("Mediana = {:0.2f}".format(CalcularMediana(alturaDeAlunos)))
```

```
print("Mediana = {:0.2f}".format(CalcularMediana(idadeAlunos)))

```

##### 4 - Escreva uma função que encontre a variância de uma lista de números reais ou inteiros. 

*Dica:* você pode reutilizar a função para cálculo da média.

```
def CalcularVariancia(lista):
  media = CalcularMedia(lista)
  variancia = 0
  for i in lista:
    variancia = variancia+(( i-media)**2)
  return variancia / (len(lista)-1)
```

##### 6 - Escreva uma função que calcule qualquer percentil de uma lista de valores. A função deve receber o $i^{o}$ índice e retornar o valor associado àquele percentil.

```
def CalcularPercentil(lista,indice):
  lista = sorted(lista)
  percentil = (indice * len(lista)) /100

  if percentil //1 == percentil:
    return lista[int(percentil-1)]
  else:
    inteiro= int(percentil)
    fracionaria = percentil - inteiro 

    return  (1 - fracionaria ) * lista[inteiro-1] + fracionaria * lista[inteiro] 
```

```
print(CalcularPercentil(listaNotas,25))
```

```
print(CalcularPercentil(alturaDeAlunos,8))
```

```
print(CalcularPercentil(alturaDeAlunos,26))
```

##### 9 - Escreva uma função que calcule a moda de uma lista de objetos (strings ou números inteiros).

```
from collections import Counter

def TotalizarObjetos(lista):
  c = Counter(lista)
  objetos  = {}
  for elemento in lista:
    if objetos.get(elemento):
      objetos[elemento] = objetos[elemento] +1
    else:
      objetos[elemento] = 1  
  return objetos

def CalcularModa(lista):
 objetos = TotalizarObjetos(lista)
 moda = max([(objeto,nome)for nome,objeto in objetos.items()])[1]
 return moda 




```

```
print(CalcularModa(listaNotas))
```

```
print(CalcularModa(nomeJoga))
```

```
print(CalcularModa(alturaDeAlunos))
```

##### 10 - Escreva uma função que calcule a proproção da moda de uma lista de objetos (strings ou números inteiros).

*Dica*: você pode reutilizar a função de cálculo da moda.

```
def CalcularProporcao(lista,moda):
  objetostotalizados = TotalizarObjetos(lista)
  totalObjetos = len(lista)
  frequenciaModa = objetostotalizados[moda]
  return frequenciaModa/totalObjetos*100
    
```

```
print(CalcularProporcao(nomeJoga,CalcularModa(nomeJoga)))
```

```
print(CalcularProporcao(alturaDeAlunos,CalcularModa(alturaDeAlunos)))
```

##### 11 - Escreva uma função que calcule a segunda moda de uma lista de objetos (strings ou números inteiros).

```
def CalcularSegundaModa(lista):
  objetos = TotalizarObjetos(lista)
  moda = CalcularModa(lista)
  objetos.pop(moda)
 
  return max([(objeto,nome)for nome,objeto in objetos.items()])[1]
```
