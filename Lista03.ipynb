##### 6 - Escreva uma função que calcule o produto matriz-vetor entre um vetor em $R^n$ e uma matriz em $R^{m\times n}$.

*Dica*: uma matriz em python pode ser representada através de uma lista de listas onde cada lista é uma linha da matriz. Exemplo:

```{python}
m = [[1,2 ],
     [3, 4]]
```

````python
def CalculeMatrizVetor(vetor,matriz):
  C= []
  for colunaMatriz   in range(len(matriz[0])):
    valor = 0
    for linhaMatriz in range(len(matriz)):
      valor = valor +vetor[linhaMatriz]*matriz[linhaMatriz][ colunaMatriz]
    C.append(valor)
  return C

````

````python
print(CalculeMatrizVetor(vetor1,C))
````

````python
print(CalculeMatrizVetor(vetor2,C))
````

````python
print(CalculeMatrizVetor(vetor3,C))
````

##### 7 - Escreva uma função que calcule o produto entre duas matrizes.

*Dica*: uma matriz pode ser interpretada como um conjunto de vetores. Para obter esta multiplicação bata obter os produtos matriz-vetor de uma matriz com os vetores da outra.

````python
def CalculoMatrizDuas(matrizA,matrizB):
  matrizResultado = []
  for vetor in  matrizA:
    matrizResultado.append(CalculeMatrizVetor(vetor,matrizB))
  return matrizResultado
  
````
